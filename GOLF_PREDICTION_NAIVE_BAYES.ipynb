```python
import pandas as pd
import numpy as np


```

```python
df=pd.read_csv('golf_df.csv')
df
```

```python
''''''''' RUNNING EDA ON DATA SET '''''

df['Outlook']=df['Outlook'].map({'sunny':2 , 'overcast':1,'rainy':0})
df
df['Temperature']=df['Temperature'].map({'hot':2 , 'mild':1,'cool':0})
df
df['Humidity']=df['Humidity'].map({'high':1 ,'normal':0})
df
df['Windy']=df['Windy'].replace({0:'False' ,1:'True'})
df
df['Play']=df['Play'].map({'no':0,'yes':1})
df['Windy'] = df['Windy'].replace({False: 0,True: 1})
df.head()
```

```python
y=df['Play']
x=df.drop(['Play'],axis=1)
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.5,random_state=42)
x
y
x

```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(x_train, y_train)
pred=model.predict(x_test)

pred

```

```python
from sklearn.metrics import r2_score
r2_score(y_test,pred)
```
